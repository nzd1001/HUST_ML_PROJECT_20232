Cosine similarity for tags

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count={}
for game_tag in games["tags"]:
    for tag in game_tag:
        if tag not in count:
            count[tag]=0
        else:
            count[tag]+=1
print(count)
```

```python
all_tags=list(set(tag for game_tag in games['tags'] for tag in game_tag if count[tag]>=100))
print(len(all_tags))
```

```python

def create_tag_vector(tags, all_tags):  
    vector = [0] * len(all_tags)
    for tag in tags:  
        if tag in all_tags:
            vector[all_tags.index(tag)] = 1
    return vector
tag_vectors=[]
for game_tag in games['tags']:
    tag_vectors.append(create_tag_vector(game_tag,all_tags))
game_tags_vec=pd.DataFrame(data=tag_vectors,columns=all_tags)
```

```python
game_tags_vec.head()
```

```python
tag_sim=cosine_similarity(game_tags_vec,game_tags_vec)
```

```python
import pickle as pkl
pkl.dump(tag_sim,open("25000_tag_sim.pkl","wb"))
```

```python
def recommend(game,sim):
    index=games[games["title"]==game].index[0]
    top_rec=sorted(list(enumerate(sim[index])),key=lambda x:x[1],reverse=True)
    for candidate in top_rec[1:11]:
        print(games.iloc[candidate[0]].title)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize("criteria",pos="v"))
```

```python
def lemmatize(text):
    y=[]
    for w in text.split():
        new_w=lemmatizer.lemmatize(w)
        y.append(new_w)
    return " ".join(y)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for w in text:
        new_w=ps.stem(w)
        y.append(new_w)
    return " ".join(y)
```
